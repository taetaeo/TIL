## 문제 02. 최댓값 찾기

***주어진 숫자 n개 중 가장 큰 숫자를 찾는 알고리즘을 만들기***

<br>

주어진 숫자 n개 중에서 가장 큰 숫자(**최댓값**)를 찾는 문제이다. 예를 들어 `17`,`92`, `18`, `33`, `58`, `7`, `33`, `42`와 같이 숫자가 여덟 개 있을 떄 최댓값은 `92`이다.
<br>

_여러 숫자를 효율적으로 다루는 데 꼭 필요한 파이썬의 **리스트** 기능_
<br>

### 1. 리스트

 `17`,`92`, `18`, `33`, `58`, `7`, `33`, `42` 와 같은 숫자 여러 개는 파이썬의 **리스트** 기능을 이용하면 쉽게 관리할 수 있다.
 **리스트(list)**는 정보 여러 개를 하나로 묶어 저장하고 관리할 수 있는 기능이다. 리스트를 만들려면 대괄호`[ ]`안에 정보 여러 개를 쉼표`,`로 구분하여 적어주면 된다.
 
 ```PYTHON

a = [5,7,8]
a #[5, 7, 9]
a[0] # 5
a[2] # 9
a[-1] # 9
len(a) # 3
```

- 첫 번째 문장의 a = [5, 7, 9]는 `5, 7, 9`라는 정수 세 개를 묶은 **리스트** 만들어 a 에 **저장**한다. 
- 두 번째 문장에서 a를 입력하면 [5, 7, 9]가 표시되면서 a가 5, 7, 9라는 정보 세 개를 묶어 놓은 리스트라고 알려준다.
- 파이썬 리스트에서 가장 주의할 점은 자료의 위치를 `1`이 아닌 `0`부터 시작이다.
    - 예를 들어, a[0] 는 리스트 a의 0번 위치 값을 의미하므로 `5`가 표시된다.
    - 마지막 값인 `9`를 얻으려면 a[2]을 사용한다. / a[3] (x)
- a[-1]의 의미는?
    - 파이썬 리스트에서 위치 번호 `-1`은 리스트의 끝에서 첫 번째 값, 즉 마지막 값을 의미한다.
        - 따라서, a[-1] = 9 라는 결과를 얻을 수 있다.
        - 예를 들어, 자료가 n개 있는 리스트 b가 있다면 첫 번째 값은 b[0]이며, 마지막 값은 b[n-1] 또는 b[-1]이 된다.
<br>

_리스트에서 자료 값의 위치 번호는 1이 아닌 0부터 시작하므로 자료가 n개일 때 마지막 값의 위치 번호는 n이 아니라 **n-1**이다._

<br>

한편, **len()** 함수를 사용하면 어떤 리스트 안에 들어 있는 **자료 개수**를 알 수 있다. 리스트 `a`에는 세 개 값인 [5, 7, 9]가 들어 있으므로 **len(a)**의 결과 3이다. 

|    함수     | 설명                                                         |
| :---------: | :----------------------------------------------------------- |
|   len(a)    | 리스트 길이(자료 개수)를 구한다.                             |
|  append(x)  | 자료 x를 리스트의 맨 뒤에 추가한다.                          |
| insert(i,x) | 리스트의 i번 위치에 x를 추가한다.                            |
|   pop(i)    | i번 위치에 있는 자료를 리스트에서 빼내면서 그 값을 함수의 결괏값으로 돌려준다. i를 지정하지 않으면 맨 마지막 값을 빼내서 돌려준다. |
|   clear()   | 리스트의 모든 자료를 지운다.                                 |
|   x in a    | 어떤 자료 x가 리스트 a 안에 있는지 확인한다.(x not in a는 반대 결과). |


_**1) len(a) 의 예**_
```python
a = []
len(a)        # 빈 리스트 이므로 0
len([1,2,3])  # 자료 개수가 세 개이므로 3
```

_**2) append(x) 의 예**_
```python
a = [1, 2, 3]
a. append(4)  # a는 [1,2,3,4]가 된다.
```

_**3) insert(i,x) 의 예**_
```python
a = [1, 2, 3]
a.insert(0,5) # 0번 위치(맨 앞)에 5를 추가한다.
              # a = [5, 1, 2, 3]이 된다.
```

_**4) pop(i) 의 예**_
```python
a = [1,2,3]
print(a.pop()) # 3이 출력되고 a = [1, 2]가 된다.
```

_**5) clear()의 예**_
```python
a = [1,2,3]
a.clear()     # a = [], 즉 빈 리스트가 된다.
```

_**6) clear()의 예**_
```python
a = [1,2,3]
2 in a     # 2가 리스트 a 안에 있으므로 True
5 in a     # 5가 리스트 a 안에 없으므로 False
5 not in a # 5가 리스트 a 안에 없으므로 True
```

### 2. 최댓값을 찾는 알고리즘

 `17`,`92`, `18`, `33`, `58`, `7`, `33`, `42` 중에서 최댓값을 찾는 알고리즘을 사람의 생각으로 정리해보자.
 

1. 첫 번째 숫자 17을 최댓값으로 기억한다. (최댓값 : 17)
2. 두 번째 숫자 92를 현재 최댓값 17과 비교한다.
    - 92는 17보다 크므로 최댓값을 92로 바꿔 기억한다. (최댓값 : 92)
3. 세 번째 숫자 18을 현재 최댓밧 92와 비교한다.
    - 18은 92보다 작으므로 지나간다. (최댓값 : 92)
4.  ~7까지 네 번째 숫자부터 일곱 번째 숫자까지 같은 과정을 반복한다.
8. 마지막 숫자 42를 현재 최댓값 92와 비교한다.
    - 42는 92보다 작으므로 지나간다. (최댓값 : 92)
9. 마지막으로 기억된 92가 주어진 숫자 중 최댓값이 된다.


### 최댓값을 구하는 알고리즘

````python
# 최댓값 구하기
# 입력 : 숫자가 n개 들어 있는 리스트
# 출력 : 숫자 n개 중 최댓값

def find_max(a):
    n = len(a)     # 입력 크기 n
    max_v = a[0]   # 리스트의 첫 번째 값을 최댓값으로 기억
    for i in range(1, n): # 1 부터 n-1까지 반복
        if a[i]>max_v:  # 이번 값이 현재까지 기억된 최댓값보다 크면
            max_v = a[i] # 최댓값을 변경
    return max_v

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_max(v))
````

### 3. 알고리즘 분석

최댓값 구하기 프로그램의 **계산 복잡도(시간 복잡도)**를 생각해보자. 입력 크기가 n일 떄, 즉, 숫자 n개 중에서 최댓값을 구할 경우 자료 개수 n은 리스트 a의 크기인 len(a)로 구할 수 있다.<br>

그렇다면 최댓값을 구하는 데 해야 하는 가장 중요한 계산은 무엇인가? 두 값 중 어느 것이 더 큰지 판단하는 **비교**이다.<br>
위의 코드에서는 `for i in range(1,n):` 반복문 안에 크기를 비교하는 판단 구문(`if a[i] > max_v:`)이 있어 자료 n개 중에서 최댓값을 찾으려면 비교를 `n-1`해야 한다. 이때 계산 복잡도는 O(n-1)이 아닌 O(n)이 된다. <br>
- 왜냐하면, n이 커질 때는 n번과 n-1의 차이가 무의미하므로 간단히 `O(n)`으로 표현한다.<br>

**계산 복잡도 O(n)**의 가장 중요한 특징은 입력 크기와 계산 시간이 대체로 **비례**한다는 것이다. 예를 들어, 숫자 10,000개 중 최댓값을 찾는 데 걸리는 시간이 10초였다면 20,000개를 입력할 때는 대략 20초가 걸릴 것으로 예상할 수 있다는 것이다.  

### 4. 응용하기

***리스트에 숫자가 n개 있을 때 가장 큰 값이 있는 위치 번호를 돌려주는 알고리즘을 만들어 보자.***

응용하여 최댓값이 리스트의 어느 위치에 있는지를 묻는 것이다. 리스트 [17, 92, 18, 33, 58, 7, 33, 42]에서 두 번째 나오는 값인 `92`가 최댓값이므로, 리스트의 1번 위치에 있으므로 결괏값은 `1`이 된다.

### 최댓값의 위치를 구하는 알고리즘

````python
# 최댓값의 위치 구하기
# 입력 : 숫자 n개 들어 있는 리스트
# 출력 : 숫자 n개 중에서 최댓값이 있는 위치 (0부터 n-1까지의 값)

def find_max_idx(a):
    n = len(a)  # 입력 크기 n
    max_idx = 0 # 리스트 중 0번 위치를 최댓값 위치로 기억
    
    for i in range(1, n):
        if a[i] > a[max_idx]: # 이번 값이 현재까지 기억된 최댓값보다 크면
            max_idx = i # 최댓값의 위치를 변경
    return max_idx

v = [17, 92, 18, 33, 58, 7, 33, 42]

print(find_max_idx(v))
````

- a[최댓값의 위치 번호] = 최댓값
- 예시 : a[1] = 92

````python
def find_max(a):
    n = len(a)
    max_v = a[0]
    for i in range(1, n):
        if a[i] > max_v:
            max_v = a[i]
    return max_v

arr = [1,2,55,99,103,204,183,4444,32521,999]
print(find_max(arr))
````

## 문제 01. 1부터 n까지의 합 구하기

`1` 부터 `10`까지의 수를 모두 더하면 ? 55 
<br>
`1` 부터 `100`까지의 수를 모두 더하면? 5050
<br>


### 1. 알고리즘의 중요 포인트

알고리즘은 어떤 문제를 풀기 위한 절차나 방법이다. 즉, 주어진 입력을 출력으로 만드는 과정을 구체적이고 명료하게 표현한 것.<br>

- **문제**<br>
알고리즘은 주어진 문제를 풀기 위한 절차나 방법이므로, 알고리즘이 있으려면 반드시 문제가 필요하다.
    - 여기서는 `1부터 n까지 연속한 숫자의 합 구하기`가 바로 문제이다.


- **입력**<br>
알고리즘은 주어진 `입력`을 `출력`으로 만드는 과정이다.이 문제에서 입력은 `n까지에 해당하는 n`이다.
    - ex) `1부터 100까지의 합을 구하시오`라면 입력 : `n`
    
    
- **출력**<br>
`n = 10`이면 1부터 10까지의 합은 `55`, `n = 100`이면 1부터 100까지의 합은 `5050`이다. 
    - `55`와 `5050`이 각각의 입력에 대한 출력이다.
    
<br>

### 2. 구체적이고 명료한 계산 과정

앞의 문제에서 무의식적으로 덧셈을 시작한 사람이 있다. 평소에 생각하는 방식을 따져 보자면, 

1. 1 더하기 2를 계산한 결과인 3을 머릿속에 기억한다.
2. 기억해 둔 3에 다음 숫자 3을 더해 6을 기억한다.
3. 기억해 둔 6에 다음 숫자 4를 더해 10을 기억한다.
4. 기억해 둔 10에 다음 숫자 5를 더해 15를 기억한다.
5. ~8. 까지 같은 과정 반복
9. 기억해 둔 45에 다음 숫자인 10을 더해 55을 기억한다.
10. 10까지 다 더했으므로 마지막에 기억된 55를 답으로 제시한다.
<br>

### 3. 1부터 n까지의 합을 구하는 알고리즘

`1`부터 `n`까지의 합을 구하는 문제를 푸는 알고리즘을 다음어 파이썬으로 프로그램을 만들어보자


![](https://github.com/ohtaekwon/TIL-Algorithms/blob/master/Modu-Algorism/img/1_1.png?raw=true)

1. 합을 기록할 변수 `s`를 만들고 `0`을 저장한다.
2. 변수 `i`를 만들어 `1`부터 `n`까지의 숫자를 1씩 증가시키며 반복한다.
3. [반복 블럭] 기존의 `s`에 `i`를 더하여 얻은 값을 다시 `s`에 저장한다.
4. 반복이 끝났을 때 `s`에 저장된 값이 결괏값이다.

<br>
위의 과정을 일련의 알고리즘으로 하나의 함수로 만들어 입력은 인자로 전달하고, 출력은 함수의 결과값(return 값)으로 만들면,<br>
`입력 → 알고리즘 → 출력`을 수행하는 과정이 된다.

<br>

----




### 1 부터 n까지 연속한 숫자의 합을 구하는 알고리즘 ①

````python
# 1 부터 n까지 연속한 숫자의 합을 구하는 알고리즘 1

# 입력 : n

# 출력 : 1부터 n까지의 숫자를 더한 값

def sum_n(n):
    s = 0  # 합을 계산할 변수
    for i in range(1, n+1): # 1부터 n까지 반복(n+1)은 제외
        s = s + i
    return s
    
    
print(sum_n(10))   # 1부터 10까지의 합(입력 : 10, 출력 55)
print(sum_n(100)) # 1부터 100까지의 합(입력 : 100, 출력 5050)
````

    합을 계산할 변수 `s`에 `0`을 넣고 첫 번째 수인 `1`을 더하는 것으로 계산을 시작하였다. 즉, `1 + 2 = 3`이 아니라 `0 + 1 = 1`이 첫 덧셈이다.

### 4. 알고리즘 분석

알고리즘은 문제를 푸는 절차나 방법이다. 그런데 어떤 문제를 푸는 방법이 한 가지만 있을까? 보통은 그렇지 않다.<br>
`1`부터 `100`까지의 합을 구하는 문제만 해도 최소 두 가지 방법이 있다.<br>
하나는 앞에서 만든 프로그램처럼 `1`부터 `100`까지의 숫자를 차례로 더하는 방법이고, 1 + 100 = 101, 2 + 99 = 101 ... 50 + 51 = 101로 `101`이 50번 나오므로 `101 * 50  = 5050`으로 **`1`부터 `n`까지의 합 공식** 인 `(n(n+1))/2`이 있다.

### 1 부터 n까지 연속한 숫자의 합을 구하는 알고리즘 ②

````python
# 1부터 n 까지 연속한 숫자의 합을 구하는 알고리즘 2

# 입력 : n

# 출력 : 1부터 n까지의 숫자를 더한 값


def sum_n(n):
    return (n*(n+1))//2  # 슬래시 두 개(//)는 정수 나눗셈을 의미한다.
        
print(sum_n(10))    # 1부터 10까지의 합 (입력 : 10 , 출력 : 55)
print(sum_n(100))   # 1부터 100까지의 합 (입력 : 100, 출력 : 5050)
````

    두 개의 방법 중 두 번째 방법이 더 좋은 방법으로 여길 것이다. 왜냐하면 첫 번째 방법은 입력 값 n이 커질수록 더 많은 덧셈을 
    반복해야하지만, 두 번째 방법은 n값의 크기와 관계없이 덧셈, 곱셈, 나눗셈을 각각 한 번씩만 하면 답을 얻을 수 있다. 
    
    이렇게 주어진 문제를 풀 떄 더 좋은 것인지 판단하는 것을 알고리즘 분석 이라고 한다.
    

### 5. 입력 크기와 계산 횟수

알고리즘은 문제를 푸는 절차나 방법이다. 그런데 어떤 문제를 푸는 방법이 한 가지만 있을까? 보통은 그렇지 않다.<br>
`1`부터 `100`까지의 합을 구하는 문제만 해도 최소 두 가지 방법이 있다.<br>
하나는 앞에서 만든 프로그램처럼 `1`부터 `100`까지의 숫자를 차례로 더하는 방법이고, 1 + 100 = 101, 2 + 99 = 101 ... 50 + 51 = 101로 `101`이 50번 나오므로 `101 * 50  = 5050`으로 **`1`부터 `n`까지의 합 공식** 인 `(n(n+1))/2`이 있다.<br>
계산하는 데 필요한 사칙연산의 횟수는 다음과 같다.
- 첫 번째 알고리즘(프로그램 1-1) : 덧셈 n번
- 두 번쨰 알고리즘(프로그램 1-2) : 덧셈, 곱셈, 나눗셈 각 한번 (총 세번)

<br>
입력 크기 n이 작을 때는 두 가지 방법이 큰 차이가 나지 않는다. 하지만 `n`이 커질수록 차이가 난다.
예를 들어 n=1000이 되면 첫 번째 알고리즘은 덧셈을 천 번 해야 하므로 세 번만 계산해도 되는 두 번쨰 알고리즘과 차이가 발생한다.<br>

보통 컴퓨터를 이용해서 계산할 떄는 입력 크기 **n**이 매우 큰 경우가 많다. 알고리즘 분석에서는 입력 크기가 매우 큰 n에 대하여 따져보는 것이 중요!!<br>
예를 들어, 우리나라 전 국민의 평균 나이를 계산하는 문제라고 하면 입력 크기 n은 우리나라 전 국민에 해당하는 50,000,000이 넘는 수이다.

### 6. 대문자 O표기법 : 계산 복잡도 표현

어떤 알고리즘이 문제를 풀기 위해 해야 하는 계산이 얼마나 복잡한지 나타낸 정도를 **계산 복잡도(complexity)**라고 한다. 계산 복잡도를 표현하는 방법에는 여러가지가 있다. 그 중 **대문자 O표기법**을 가장 많이 사용한다. (대문자 O표기법은 `빅 오`표기법이라고도 부른다.)<br>

첫 번째 알고리즘의 계산 복잡도를 `O(n)`이라고 표현한다. 필요한 계산 횟수가 입력 크기에 **정비례** 할때는 `O(n)`이라고 표현한다.
입력 크기 `n`에 따라 덧셈을 두 번씩 하는 알고리즘이 있다면 어떨까? 얼핏 생각하면 O(2n)으로 표현할 것 같지만 그렇지 않다.<br>
마찬가지로 그냥 `O(n)`으로 표현한다. 대문자 O 표기법은 알고리즘에서 필요한 계산 횟수를 정확한 숫자로 표현하는 것이 아니라 **입력 크기와의 관계**로 표현한다.<br>
예를 들어, n이 10에서 20으로 `2배`가 될 때 `2n` 역시 20에서 40으로 `2배`가 된다. 이처럼 필요한 계산 횟수가 입력 크기 `n`과 **정비례**하면 모두 O(n)으로 표기한다.
<br>
<br>
두 번째 알고리즘은 입력 크기 n과 무관하게 사칙연산을 세 번해야 한다. 이때 알고리즘의 계산 복잡도는 `O(1)`로 표현한다. 왜 O(3)이 아닐까? 앞에서 O(2n)을 O(n)으로 표현하는 것과 같은 원리이다. 계산 시간이 더 늘어나지 않는다면 모두 `O(1)`로 표기한다.<br>
**대문자 O표기법**은 알고리즘의 대략적인 **성능**을 표시하는 방법이다. 따라서 세밀한 계산 횟수가 소요 시간을 표현하기보다는 **입력 크기 n**과 필요한 계산횟수와의 **관계**에 더 초점을 둔다.<br>
- **O(n)** : 필요한 계산 횟수가 입력 크기 n과 비례할 때
- **O(n)** : 필요한 계산 횟수가 입력 크기 n과 무관할 때

### 1-1. _1부터 n까지 연속한 숫자의 제곱의 합을 구하는 프로그램을 for 반복문으로 만들어 보시오_
(예를 들어, n=10 이라면 1^2 + 2^2 + 3^2 + ... + 10^2 = 385를 계산하는 프로그램이다.)

````python
# 연속한 숫자 제곱의 합을 구하는 알고리즘
# 입력: n
# 출력: 1부터 n까지 연속한 숫자의 제곱을 더한 합

def sum_sq(n):
    s=0
    for i in range(1, n+1):
        s=  s + (i*i)
    return s

print(sum_sq(10))   # 1부터 10까지 제곱의 합(입력:10, 출력:385)
print(sum_sq(100))  # 1부터 100까지 제곱의 합(입력:100, 출력:338350)
````

### 1-3. _1부터 n까지 연속한 숫자의 제곱의 합을 구하는 공식은 {n(n+1)(2n+1)}/6 으로 알려져있다. for 반복문 대신 이 공식을 이용하면 알고리즘의 계산 복잡도는 O(1)과 O(n)중 무엇일까?_

````python
def sum_n2(n):
    return (n*(n+1)*(2*n+1))//6

print(sum_n2(10))
````

````python
# 연속한 숫자 제곱의 합 구하기 알고리즘
# 입력: n
# 출력: 1부터 n까지 연속한 숫자의 제곱을 더한 합

def sum_sq(n):
    return n * (n + 1) * (2 * n + 1) // 6

print(sum_sq(10))   # 1부터 10까지 제곱의 합(입력:10, 출력:385)
print(sum_sq(100))  # 1부터 100까지 제곱의 합(입력:100, 출력:338350)
````
